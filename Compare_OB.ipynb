```python
import numpy as np
import matplotlib.pyplot as plt
import json
import os
```

```python
file = open('run1724/Degeling_7th/PSD_reduced_7th.pysav','rb')
roa_7th = np.load(file)
mltoa_7th = np.load(file)
psd_7th = np.load(file)
houra_7th = np.load(file)
xK0_7th = np.load(file)
xmm0_7th = np.load(file)
lat0_7th = np.load(file)
file.close
```

```python
file = open('run1724/OB_0000s/PSD_reduced_0000s.pysav','rb')
roa_0000s = np.load(file)
mltoa_0000s = np.load(file)
psd_0000s = np.load(file)
houra_0000s = np.load(file)
xK0_0000s = np.load(file)
xmm0_0000s = np.load(file)
lat0_0000s = np.load(file)
file.close
```

```python
file = open('run1724/OB_5400s/PSD_reduced_5400s.pysav','rb')
roa_5400s = np.load(file)
mltoa_5400s = np.load(file)
psd_5400s = np.load(file)
houra_5400s = np.load(file)
xK0_5400s = np.load(file)
xmm0_5400s = np.load(file)
lat0_5400s = np.load(file)
file.close
```

```python
z_7th = np.zeros([len(houra_7th)*len(lat0_7th)*len(xmm0_7th)]).reshape(len(houra_7th),len(lat0_7th),len(xmm0_7th))
for iTime in range(len(houra_7th)):
    for iMu in range(len(xmm0_7th)):
        z_7th[iTime,:,iMu]=\
            np.log(psd_7th[iTime,:,0,iMu]/psd_7th[0,:,0,iMu])

z_0000s = np.zeros([len(houra_0000s)*len(lat0_0000s)*len(xmm0_0000s)]).reshape(len(houra_0000s),len(lat0_0000s),len(xmm0_0000s))
for iTime in range(len(houra_0000s)):
    for iMu in range(len(xmm0_0000s)):
        z_0000s[iTime,:,iMu]=np.log(psd_0000s[iTime,:,0,iMu]/psd_0000s[0,:,0,iMu])

z_5400s= np.zeros([len(houra_5400s)*len(lat0_5400s)*len(xmm0_5400s)]).reshape(len(houra_5400s),len(lat0_5400s),len(xmm0_5400s))
for iTime in range(len(houra_5400s)):
    for iMu in range(len(xmm0_5400s)):
        z_5400s[iTime,:,iMu]=np.log(psd_5400s[iTime,:,0,iMu]/psd_5400s[0,:,0,iMu])
```

```python
f, axs = plt.subplots(11, 3, figsize=(20, 60), sharex = False, sharey = False)

#axs[0,0].axis([1.5, 6.0, 1., 10.])
print(type(f))
print(axs[1])
#ax1=f.add_subplot(121)
#ax2=f.add_subplot(122)

lat_in_radians = np.radians(lat0_7th)
Lshell_7th = np.cos(lat_in_radians)**-2

#print(np.transpose(z).shape)
lat_in_radians = np.radians(lat0_0000s)
#print(lat_in_radians)
Lshell_0000s = np.cos(lat_in_radians)**-2
#print(Lshell_0000s)
#Lshell = 1

#print(np.transpose(z).shape)
lat_in_radians = np.radians(lat0_5400s)
#print(lat_in_radians)
Lshell_5400s = np.cos(lat_in_radians)**-2
#print(Lshell_5400s)
#Lshell = 1

print(xmm0_0000s[24:35]*100.)
cont_levarr = np.linspace(-20,20,num=41)
#print(cont_levarr)

vmin_arr = \
    np.array([ -00.1, -00.1, -00.1, -00.1, -00.25, -00.5, -01.0, -01.0, -02., -02., -02. ])
vmax_arr = \
    np.array([  00.1,  00.1,  00.1,  00.1,  00.25,  00.5,  01.5,  02.5,  08.,  15.,  15. ])

for iMu in range(24,35):
    iRow = iMu % 24

    ax_l = axs[iRow, 0]
    ax_l.axis([1.5, 6.0, 1., 10.])
    ax_l.set_ylabel('L')
    ax_l.set_xlabel('Simulation Time [hours]')
    ax_l.title.set_text(str(int(xmm0_7th[iMu]*100.))+' MeV/G')
    cf_l = ax_l.pcolormesh(houra_7th, Lshell_7th, np.transpose(z_7th[:,:,iMu]),\
                           vmin=vmin_arr[iMu-24],vmax=vmax_arr[iMu-24])
    f.colorbar(cf_l, ax=ax_l)

    ax_c = axs[iRow, 1]
    ax_c.axis([1.5, 6.0, 1., 10.])
    ax_c.set_ylabel('L')
    ax_c.set_xlabel('Simulation Time [hours]')
    ax_c.title.set_text(str(int(xmm0_0000s[iMu]*100.))+' MeV/G')
    cf_c = ax_c.pcolormesh(houra_0000s, Lshell_0000s, np.transpose(z_0000s[:,:,iMu]),\
                           vmin=vmin_arr[iMu-24],vmax=vmax_arr[iMu-24])
    f.colorbar(cf_c, ax=ax_c)

    ax_r = axs[iRow, 2]
    ax_r.axis([1.5, 6.0, 1., 10.])
    ax_r.set_ylabel('L')
    ax_r.set_xlabel('Simulation Time [hours]')
    ax_r.title.set_text(str(int(xmm0_5400s[iMu]*100.))+' MeV/G')
    cf_r = ax_r.pcolormesh(houra_5400s, Lshell_5400s, np.transpose(z_5400s[:,:,iMu]),\
                           vmin=vmin_arr[iMu-24],vmax=vmax_arr[iMu-24])
    f.colorbar(cf_r, ax=ax_r)

f.savefig("Initialization_Comparison.png")
#axs[0,0].set_title('Degeling initialized at t=00:00')
#axs[0,1].set_title('Degeling initialized at t=01:30')
```
